```python
import pandas as pd
import glob
import os
# read file from wunderground.com
path = r'D:\bootcamp\week_20_Final_Project\BIke sharing\weather\Wunderground'                    
all_files = glob.glob(os.path.join(path, "*.csv"))

# read each csv file and concatenate it into one dataframe
df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
concatenated_df
```

```python
#delete unnecessary rows
concatenated_df = concatenated_df.loc[concatenated_df["W_Max_Dp"] != 'Max']
```

```python
concatenated_df
```

```python
#slice Year column data 
concatenated_df['W_Year'] = concatenated_df['W_Year'].str.slice(2,7)
concatenated_df
```

```python
#Mapping Month Name to number
Month_mapping = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
concatenated_df.W_Month= concatenated_df.W_Month.map(Month_mapping)
```

```python
#Change datatype 
concatenated_df['W_Year']=concatenated_df['W_Year'].astype(str).astype('int64')
concatenated_df['W_Day']=concatenated_df['W_Day'].astype(str).astype('int64')
concatenated_df['W_Max_Dp']=concatenated_df['W_Max_Dp'].astype(str).astype('float64')
concatenated_df['W_Avg_dp']=concatenated_df['W_Avg_dp'].astype(str).astype('float64')
concatenated_df['W_Min_Dp']=concatenated_df['W_Min_Dp'].astype(str).astype('float64')
concatenated_df['W_Max_humid']=concatenated_df['W_Max_humid'].astype(str).astype('float64')
concatenated_df['W_Avg_humid']=concatenated_df['W_Avg_humid'].astype(str).astype('float64')
concatenated_df['W_Min_humid']=concatenated_df['W_Min_humid'].astype(str).astype('float64')
concatenated_df['W_Max_wind']=concatenated_df['W_Max_wind'].astype(str).astype('float64')
concatenated_df['W_Avg_wind']=concatenated_df['W_Avg_wind'].astype(str).astype('float64')
concatenated_df['W_Min_wind']=concatenated_df['W_Min_wind'].astype(str).astype('float64')

concatenated_df.dtypes
```

```python
# Change Metric from Fahrenheit to Celcius 
concatenated_df['W_Max_Dp'] = (concatenated_df['W_Max_Dp'] - 32) * 5 / 9
concatenated_df['W_Avg_dp'] = (concatenated_df['W_Avg_dp'] - 32) * 5 / 9
concatenated_df['W_Min_Dp'] = (concatenated_df['W_Min_Dp'] - 32) * 5 / 9
#Change Metric from mph to kmh
concatenated_df['W_Max_wind']=concatenated_df['W_Max_wind']*1.60934
concatenated_df['W_Avg_wind']=concatenated_df['W_Avg_wind']*1.60934
concatenated_df['W_Min_wind']=concatenated_df['W_Min_wind']*1.60934
```

```python
#rounding Data 
decimals = 1    
concatenated_df['W_Max_Dp'] = concatenated_df['W_Max_Dp'].apply(lambda x: round(x, decimals))
concatenated_df['W_Avg_dp'] = concatenated_df['W_Avg_dp'].apply(lambda x: round(x, decimals))
concatenated_df['W_Min_Dp'] = concatenated_df['W_Min_Dp'].apply(lambda x: round(x, decimals))
concatenated_df['W_Max_wind'] = concatenated_df['W_Max_wind'].apply(lambda x: round(x, decimals))
concatenated_df['W_Avg_wind'] = concatenated_df['W_Avg_wind'].apply(lambda x: round(x, decimals))
concatenated_df['W_Min_wind'] = concatenated_df['W_Min_wind'].apply(lambda x: round(x, decimals))
concatenated_df.head()
```

```python
#Make sure there is no missing data for the day
concatenated_df.groupby(['W_Year', 'W_Month']).size().reset_index(name='counts')
```

```python
# read file from meteostat.net
meteo_df=pd.read_csv('Meteo_Data/export.csv')
```

```python
#check dataframe
meteo_df
```

```python
#Slice date and make new column (Year,Month,Day)
meteo_df['Year'] = pd.DatetimeIndex(meteo_df['date']).year
meteo_df['Month'] = pd.DatetimeIndex(meteo_df['date']).month
meteo_df['Day'] = pd.DatetimeIndex(meteo_df['date']).day
#drop column date
meteo_df = meteo_df.drop('date', 1)
meteo_df.head()
```

```python
#Make sure there is no missing data for the day
meteo_df.groupby(['Year', 'Month']).size().reset_index(name='counts')
```

```python
# drop wspd,tsun
meteo_df = meteo_df.drop(['wspd','tsun'], 1)
meteo_df.head()
```

```python
#change column name to be more easier to understand
meteo_df.columns = ['W_Temp_Avg','W_Temp_Min','W_Temp_Max','W_Precipitation','W_Snow_Depth','W_Wind_Dir','W_Wind_Gust','W_Air_Pressure','W_Year','W_Month','W_Day']

#change column name to be more clean
#concatenated_df.columns['W_Year','W_Month','W_Day','W_Max_Dp','W_Avg_Dp','W_Min_Dp','W_Max_Wind','W_Avg_Wind','W_Min_Wind','W_Max_Humid','W_Avg_Humid','W_Min_Humid']

#Merge meteo df and concatenated_df
df = pd.merge(meteo_df,concatenated_df,on=['W_Year', 'W_Month','W_Day'], how='left')


df.head()
```

```python
#change column order
df = df[['W_Year','W_Month','W_Day','W_Temp_Max','W_Temp_Avg','W_Temp_Min','W_Max_wind','W_Avg_wind','W_Min_wind','W_Wind_Gust','W_Air_Pressure','W_Snow_Depth','W_Max_humid','W_Avg_humid','W_Min_humid','W_Max_Dp','W_Avg_dp','W_Min_Dp']]

df.head()
```

```python
#Make sure there is no missing data for the day
df.groupby(['W_Year', 'W_Month']).size().reset_index(name='counts')
```

```python
df.to_csv('result/result.csv',index=False)
```
