```python
import pandas as pd
import glob
import os
# read file from Bike share csv
path = r'D:\bootcamp\week_20_Final_Project\BIke sharing\Bike data'                    
all_files = glob.glob(os.path.join(path, "*.csv"))

# read each csv file and concatenate it into one dataframe
df_from_each_file = (pd.read_csv(f,encoding='latin-1') for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
concatenated_df.head()
```

```python
#delete na row

concatenated_df.dropna(subset = ['User Type'], inplace=True)#249 rows
```

```python
concatenated_df.dtypes
```

```python
# drop if there is na data and check if there is still na data

concatenated_df.dropna(subset = ['Trip  Duration'], inplace=True)
concatenated_df[concatenated_df['Trip  Duration'].isna()]
```

```python
# change datetype from datetime to date
concatenated_df['Date']=pd.to_datetime(concatenated_df['Start Time']).dt.date
```

```python
#check start station id =na 
concatenated_df[concatenated_df['Start Station Id'].isna()]
```

```python
# copy to another dataframe 
clean_df2=concatenated_df
```

```python
#fill End station=start station if end station id =na

clean_df2['End Station Id'] = clean_df2['End Station Id'].fillna(clean_df2['Start Station Id'])

clean_df2
```

```python
#check end station id= na 
clean_df2[clean_df2['End Station Id'].isna()]
```

```python
# Convert station id to float to make all of the column has same datatype
clean_df2['Start Station Id']=clean_df2['Start Station Id'].astype(str).astype('float64')
clean_df2['End Station Id']=clean_df2['End Station Id'].astype(str).astype('float64')
```

```python
#check unique value on start station 
clean_df2['Start Station Id'].nunique()
```

```python
#check unique value for end station 
clean_df2['End Station Id'].nunique()
```

```python
# open station file 
import json
with open('station_information.json', 'r') as f:
    data = json.load(f)['data']['stations']
Station_info_df = pd.DataFrame(data)
```

```python
#covert station id type to float
Station_info_df['station_id']=Station_info_df['station_id'].astype(str).astype('float64')
```

```python
# pick necessary column 
Station_info_df=Station_info_df[['station_id','name','lat','lon']]
```

```python
#Merge to clean df to make sure all the station name will be the same 
Merged_station_data_df=pd.merge(clean_df2,Station_info_df,left_on='Start Station Id', right_on='station_id',how='left')
Merged_station_data_df
```

```python
#fillna if there is no  start station name in raw data using station name in Station_info_df
Merged_station_data_df['Start Station Name'] = Merged_station_data_df['Start Station Name'].fillna(Merged_station_data_df['name'])
Merged_station_data_df[Merged_station_data_df['Start Station Name'].isna()]
```

```python
#Check if there is any start station in raw data that doesn't have any info in Station_info_df 
Start_station_missing_df=Merged_station_data_df[Merged_station_data_df['name'].isna()]
Start_station_missing_df= Start_station_missing_df.drop_duplicates(['Start Station Id','Start Station Name'])[['Start Station Id','Start Station Name']]
Start_station_missing_df
```

```python
#change column name for end station
Station_info_df.columns=['end_stat_id','end_name','end_lat','end_lon']
Station_info_df
```

```python
# merge to raw data using end station as key 
Merged_station_data_df=pd.merge(Merged_station_data_df,Station_info_df,left_on='End Station Id', right_on='end_stat_id',how='left')
Merged_station_data_df
```

```python
#fillna if there is no  end station name in raw data using station name in station_df
Merged_station_data_df['End Station Name'] = Merged_station_data_df['End Station Name'].fillna(Merged_station_data_df['end_name'])
Merged_station_data_df[Merged_station_data_df['End Station Name'].isna()]
```

```python
#drop duplicates for end station
End_station_missing_df=Merged_station_data_df[Merged_station_data_df['end_name'].isna()|Merged_station_data_df['End Station Name'].isna()]
End_station_missing_df= End_station_missing_df.drop_duplicates(['End Station Id','End Station Name'])[['End Station Id','End Station Name']]
End_station_missing_df
```

```python
#all location missing from start station and end station 
End_station_missing_df.columns=['Start Station Id','Start Station Name']
All_Location = Start_station_missing_df.append(End_station_missing_df, ignore_index=True)
All_Location
```

```python
#delete duplicate nan
All_Location = All_Location[~All_Location['Start Station Id'].duplicated(keep=False) | All_Location[['Start Station Name']].notnull().any(axis=1)]
All_Location
```

```python
#drop duplicates
All_Location= All_Location.drop_duplicates(['Start Station Id','Start Station Name'])[['Start Station Id','Start Station Name']]
All_Location
```

```python
#check how many station is double
All_Location.groupby(['Start Station Id']).count()
```

```python
# sort by station id
All_Location=All_Location.sort_values(by=['Start Station Id'])
All_Location
```

```python
#save to csv
All_Location.to_csv('result/missing_station.csv',index=False)
```
