```python
#import the csv files
Weather_df = pd.read_csv('Weather/weather_data_clean.csv')
Bike_data_df= pd.read_csv('Bike_data/Bike_data.csv')
```

```python
#change column name to match column name in database
Weather_df.columns = ['w_date','w_temp_max','w_temp_avg','w_temp_min','w_max_wind','w_avg_wind','w_min_wind','w_wind_gust','w_air_pressure','w_snow_depth','w_max_humid','w_avg_humid','w_min_humid','w_max_dp','w_avg_dp','w_min_dp']
Weather_df
```

```python
#change column name to match table in database
Bike_data_df.columns=['b_date','b_counts_trips','b_counts_bike','b_trip_duration','b_counts_member_annual','b_counts_member_casual']
Bike_data_df
```

```python
Weather_df=Weather_df.fillna(0)
Bike_data_df=Bike_data_df.fillna(0)
```

```python
Weather_df
```

```python
# Import Modules and Create the Database Engine
from sqlalchemy import create_engine
import psycopg2
from config import db_password

db_string = f"postgresql://postgres:{db_password}@localhost/Bike_sharing"
```

```python
# Create the database engine 
engine = create_engine(db_string)
```

```python
#import bike data
engine.execute('TRUNCATE tbl_bike_data RESTART IDENTITY;')
Bike_data_df.to_sql('tbl_bike_data',con=engine, if_exists='append',index=False)
```

```python
#import weather data
engine.execute('TRUNCATE tbl_weather_data RESTART IDENTITY;')
Weather_df.to_sql(name="tbl_weather_data", con=engine, if_exists='append',index=False)   
```
